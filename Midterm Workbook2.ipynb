```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

housing = pd.read_excel('/content/drive/MyDrive/Housing Data workbook 2.xlsx', sheet_name='DA_-_housing-price-data-0404201')
housing.head()

#select basement=0

no_basement_houses = housing[housing['TotalBsmtSF'] == 0]

#group by year and calc avg

avg_price_by_year = no_basement_houses.groupby('YrSold')['SalePrice'].mean()

print(avg_price_by_year)
```

```python
#select basement>0

basement_houses = housing[housing['TotalBsmtSF'] > 0]

#group by year and calc avg

avg_price_by_year = basement_houses.groupby('YrSold')['SalePrice'].mean()

print(avg_price_by_year)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

housing_2= pd.read_excel('/content/drive/MyDrive/AmesIOWABMSQFTMidterm.xlsx')
housing_2.head()
sb.regplot(data=housing_2, x="TotalBsmtSF", y="SalePrice", color='#7c93c2')

plt.title('Home Sales in 2006 Based on Basement Area')
plt.xlabel('Basement Area (Sq. Ft)')
plt.ylabel('Sale Price ($)')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_excel('/content/drive/MyDrive/Housing Data workbook 2.xlsx', sheet_name='DA_-_housing-price-data-0404201')

df['Has_Basement'] = df['TotalBsmtSF'].apply(lambda x: 1 if x > 0 else 0)
yearly_sales = pd.pivot_table(df, index='YrSold', columns='Has_Basement', values='SalePrice', aggfunc='mean')

plt.plot(yearly_sales.index, yearly_sales[1], label='With Basement')
plt.plot(yearly_sales.index, yearly_sales[0], label='Without Basement')
plt.xlabel('Year')
plt.ylabel('Average Sale Price ($)')
plt.title('Average Sale Price by Year and Basement')
plt.legend()
plt.show()
```
